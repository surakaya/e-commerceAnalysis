**Madencilik Teknikleri**

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Kategorik sütunu sayısal hale getirmek için LabelEncoder kullanıyoruz
le = LabelEncoder()
data['uyelik_turu'] = le.fit_transform(data['uyelik_turu'])

# Özellikler ve hedef değişken
X = data.drop('memnuniyet_duzeyi', axis=1)
y = data['memnuniyet_duzeyi']

# Eğitim ve test verilerini ayıralım
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Karar Ağaçları Modeli
model = DecisionTreeClassifier()

# Hiperparametre ızgarası
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10]
}

# Hiperparametre optimizasyonu ve en iyi modeli bulma
grid_search = GridSearchCV(model, param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# En iyi parametreler
print(f"En iyi parametre: {grid_search.best_params_}")

# Modeli değerlendirelim
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print(f"Decision Tree Model Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier


# Kategorik sütunu sayısal hale getirmek için LabelEncoder kullanıyoruz
le = LabelEncoder()
data['uyelik_turu'] = le.fit_transform(data['uyelik_turu'])

# Özellikler ve hedef değişken
X = data.drop('memnuniyet_duzeyi', axis=1)
y = data['memnuniyet_duzeyi']

# Eğitim ve test verilerini ayıralım
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Rastgele Ormanlar Modeli
model = RandomForestClassifier()

# Hiperparametre ızgarası
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10]
}

# Hiperparametre optimizasyonu ve en iyi modeli bulma
grid_search = GridSearchCV(model, param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# En iyi parametreler
print(f"En iyi parametre: {grid_search.best_params_}")

# Modeli değerlendirelim
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print(f"Random Forest Model Doğruluğu: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))
```
